```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
def gaussian_y(xi, mean, sigma):
    const = 1/(((2*math.pi)**0.5)*sigma)
    y_var = (math.e)**((-0.5*((sigma)**2))*((xi-mean)**2))
    y = const*y_var
    return y


def get_likelihood_prob(x_arr, mean, sigma=1):
    product = 1
    for xi in x_arr:
        product = product*gaussian_y(xi, mean, sigma)
    return product
```

```python
mean = 5
variance = 1
sigma = math.sqrt(variance)
const = 1/(((2*math.pi)**0.5)*sigma)
x_arr = random.normal(mean, sigma, 10)
y_arr = [gaussian_y(x, mean, sigma) for x in x_arr]
plt.scatter(x_arr, y_arr)
plt.savefig("datapoints.png")
plt.title("Datapoints")
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca()
means = [i for i in range(1, 11)]
likelihood_arr = []
for mean in means:
    likelihood_arr.append(get_likelihood_prob(x_arr, mean))
ax.scatter(means, likelihood_arr)
ax.plot(means, likelihood_arr, color = "orange")
ax.relim()
ax.autoscale_view()
plt.xlabel("Means")
plt.ylabel("Likelihood")
plt.title("Likelihood Probability for different means")
plt.savefig("Likelihood.png")
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca()
log_likelihood = [math.log(num) for num in likelihood_arr]
ax.scatter(means, likelihood_arr)
ax.plot(means, log_likelihood , color = "violet")
ax.relim()
ax.autoscale_view()
plt.xlabel("Means")
plt.ylabel("Log Likelihood")
plt.title("Log Likelihood for different means")
plt.savefig("log likelihood.png")
plt.show()
```
